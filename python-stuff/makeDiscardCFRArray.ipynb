```python
disc_or_no = []
eml = []
deadwood = []
cardid = []
last_card_id = 0

for line in in_file:
    stripped_line = line.strip()
    str_arr = stripped_line.split(" - ")
    
    if "Discard" in str_arr[0]:
        disc_or_no.append(1 if float(str_arr[-1]) >= 0.500 else 0)
    else:
        disc_or_no.append(1 if float(str_arr[-1]) >= 0.500 else 0)
        eml.append(disc_or_no)
        disc_or_no = []
        
        if "L" in str_arr[1]:
            deadwood.append(eml)
            eml = []
        
            if int(str_arr[2]) == 80:
                cardid.append(deadwood)
                deadwood = []

# Somethings broke in the file output, just assume its okay to discard
# card id 51 .. worst case scenerio, its just as good as the old method :)
dummy_eml = [[1,1],[1,1],[1,1]]
for i in range(38, 81):
    deadwood.append(dummy_eml)

cardid.append(deadwood)
```

```python
deadwood = 0
card = 0
stageCount = 0

str_builder = "{"
for cardinfo in cardid:
    str_builder += "_____________________________ START CARD: " + str(card) + "______________________\n"
    card += 1
    str_builder += "{"
    for deadwood_val in cardinfo:
        deadwood += 1
        str_builder += "{"
        for stage in deadwood_val:
            stageCount += 1
            str_builder += "{"
            for i in range (0, 2):
                str_builder += str(stage[i])
                if (i == 0):
                    str_builder += ", "
            if stageCount == 3:
                str_builder += "} \n"
                stageCounter = 0
            else:
                str_builder += "}, \n"
        if deadwood == 80:
            str_builder += "} \n"
        else:
            str_builder += "}, \n"
    if card == 52:
        str_builder += "} \n"
    else:
        str_builder += "}, \n"
    str_builder += "_____________________________ END CARD: " + str(card - 1) + "______________________\n"
str_builder += "}"

#print(str_builder)
            
```
